```python
import nltk
import matplotlib.pyplot as plt
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
nltk.download('punkt')

# Read the text file
with open('youtube_comments.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Initialize the sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a list of positive emojis
positive_emojis = ["👌", "👍", "😍", "🥰", "☺️", "💖", "❤","👋", "🔥"]

# Initialize counters for sentiment categories
positive_count = 0
negative_count = 0
neutral_count = 0

# Perform sentiment analysis on each sentence
for line in lines:
    # Check if any positive emojis are present in the line
    has_positive_emoji = any(emoji in line for emoji in positive_emojis)
    
    if has_positive_emoji:
        positive_count += 1
        sentiment_label = 'Positive'
    else:
        # Calculate sentiment scores
        sentiment_scores = sid.polarity_scores(line)

        # Determine the sentiment label based on the compound score
        if sentiment_scores['compound'] >= 0.05:
            positive_count += 1
            sentiment_label = 'Positive'
        elif sentiment_scores['compound'] <= -0.05:
            negative_count += 1
            sentiment_label = 'Negative'
        else:
            neutral_count += 1
            sentiment_label = 'Neutral'

    print("Line:", line.strip())
    print("Sentiment:", sentiment_label)
    print()

# Print the sentiment counts
print("Positive count:", positive_count)
print("Negative count:", negative_count)
print("Neutral count:", neutral_count)

# Visualize the sentiment counts
labels = ['Positive', 'Negative', 'Neutral']
counts = [positive_count, negative_count, neutral_count]

plt.bar(labels, counts)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Analysis Results (YouTube)')

# Add count values as annotations to the bars
for i, count in enumerate(counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()


# Visualize the sentiment counts

# Pie chart
plt.figure(figsize=(6, 6))
colors = ['lightgreen', 'red', 'gray']
plt.pie([positive_count, negative_count, neutral_count], labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Sentiment Analysis Results_YouTube (Pie Chart)')

# Add total count as an annotation
total_count = sum(counts)
plt.text(0, -1.2, "Total Count: {}".format(total_count), ha='center', va='top')

plt.show()
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the YouTube comments from the text file
with open("youtube_comments.txt", "r", encoding="utf-8") as file:
    comments = file.read().splitlines()

# Preprocess the comments
nltk.download("stopwords")
nltk.download("punkt")
stop_words = set(stopwords.words("english"))
processed_comments = []
for comment in comments:
    tokens = word_tokenize(comment.lower())
    filtered_tokens = [token for token in tokens if token.isalpha() and token not in stop_words]
    processed_comments.append(" ".join(filtered_tokens))

# Perform topic modeling
vectorizer = CountVectorizer()
comment_matrix = vectorizer.fit_transform(processed_comments)
lda_model = LatentDirichletAllocation(n_components=2, random_state=42)
topics = lda_model.fit_transform(comment_matrix)
topic_labels = ["Topic {}".format(i) for i in range(lda_model.n_components)]

# Display topic distribution for each comment
topic_results = pd.DataFrame(topics, columns=topic_labels)
print(topic_results.head())

# Create a word cloud visualization
word_cloud_text = " ".join(processed_comments)
word_cloud = WordCloud(width=800, height=400, background_color="white").generate(word_cloud_text)

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from datetime import datetime

nltk.download('vader_lexicon')
nltk.download('punkt')

# Read the text file
with open('youtube_comments.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Define the event date
event_date = datetime(2023, 2, 13)

# Initialize the sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Define a list of positive emojis
positive_emojis = ["👌", "👍", "😍", "🥰", "☺️", "💖", "❤", "👋", "🔥"]

# Initialize counters for sentiment categories
positive_count_before = 0
negative_count_before = 0
neutral_count_before = 0

positive_count_after = 0
negative_count_after = 0
neutral_count_after = 0

# Perform sentiment analysis on each sentence
for line in lines:
    try:
        # Extract the date from the line
        comment_date = datetime.strptime(line.strip().split(',')[0], "%Y-%m-%d")
    except ValueError:
        # Skip the line if the date format is invalid
        continue

    # Calculate sentiment scores
    sentiment_scores = sid.polarity_scores(line)

    # Determine the sentiment label based on the compound score
    if sentiment_scores['compound'] >= 0.05:
        sentiment_label = 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        sentiment_label = 'Negative'
    else:
        sentiment_label = 'Neutral'
    
    # Compare the date with the event date
    if comment_date < event_date:
        if any(emoji in line for emoji in positive_emojis):
            positive_count_before += 1
        elif sentiment_label == 'Positive':
            positive_count_before += 1
        elif sentiment_label == 'Negative':
            negative_count_before += 1
        else:
            neutral_count_before += 1
    else:
        if any(emoji in line for emoji in positive_emojis):
            positive_count_after += 1
        elif sentiment_label == 'Positive':
            positive_count_after += 1
        elif sentiment_label == 'Negative':
            negative_count_after += 1
        else:
            neutral_count_after += 1

    print("Line:", line.strip())
    print("Sentiment:", sentiment_label)
    print()

# Print the sentiment counts before and after the event date
print("Sentiment counts before the event (13th Feb 2023):")
print("Positive count:", positive_count_before)
print("Negative count:", negative_count_before)
print("Neutral count:", neutral_count_before)
print()

print("Sentiment counts after the event (13th Feb 2023):")
print("Positive count:", positive_count_after)
print("Negative count:", negative_count_after)
print("Neutral count:", neutral_count_after)
```
